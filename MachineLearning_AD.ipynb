```python
import pandas as pd
url='Advertising.csv'
data=pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation =data.corr()
print(correlation)
```

```python
data.plot(kind='scatter',x='sales',y='TV',figsize=(9,9),color='black')
```

```python
from sklearn import linear_model
model=linear_model.LinearRegression()
#Ham day hoc
model.fit(X=pd.DataFrame(data['sales']),y=data['TV'])
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['sales']),y=data['TV'])
data.plot(kind='scatter',x='sales',y='TV',figsize=(9,9),color='black')
y_pred=model.predict(X=pd.DataFrame(data['sales']))
plt.plot(data['sales'],y_pred,color='blue')

```

```python
model=linear_model.LinearRegression()
X= pd.DataFrame([data['sales'],data['sales']**2]).T
model.fit(X,y=data['TV'])
print('Intercept cua mon hinh',model.intercept_)
print('Coffient cua mo hinh',model.coef_)
print('Do chinh xac cua mo hinh',model.score(X,y=data['TV']))
```

```python
import numpy as np
range=np.arange(1.5,5.5,0.1)
data.plot(kind='scatter',x='sales',y='TV',figsize=(9,9),color='black')
predictor=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```

```python
#Tạo ra mô hình mới bậc 10
model = linear_model.LinearRegression()
X = pd.DataFrame([data['sales'],data['sales']**2,data['sales']**3,data['sales']**4,data['sales']**5,data['sales']**6,data['sales']**7,data['sales']**8,data['sales']**9,data['sales']**10]).T
model.fit(X,y=data['TV'])
print('Intercept của mô hình',model.intercept_)
print('Coeffient của mô hình',model.coef_)
print('Độ chính xác của mô hình',model.score(X,y=data['TV']))

range = np.arange(1.5,20,0.1)
data.plot(kind='scatter',x ='sales',y='TV',figsize=(9,9),color='black')
predictor = pd.DataFrame([range,range**2,range**3,range**4,range**5,range**6,range**7,range**8,range**9,range**10]).T
y_pred = model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
