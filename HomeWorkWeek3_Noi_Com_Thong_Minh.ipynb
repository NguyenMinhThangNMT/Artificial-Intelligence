```python
import numpy as np
import skfuzzy as Fuzz
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
#Khai bao tong quat
time= ctrl.Antecedent(np.arange(0,50,1),'time')
rice= ctrl.Antecedent(np.arange(0.2,1,0.01),'rice')
temp= ctrl.Consequent(np.arange(0,100,1),'temp')
x=np.arange(30,80,0.1)
```

```python
#Thiet lap thong so thoi gian
time ['VF'] =Fuzz.trimf(time.universe,[0,0,5])
time ['F'] =Fuzz.trimf(time.universe,[4,5,10])
time ['AV'] =Fuzz.trimf(time.universe,[9,10,20])
time ['S'] =Fuzz.trimf(time.universe,[18,25,35])
time ['VS'] =Fuzz.trimf(time.universe,[30,40,50])
time ['VF'].view()
```

```python
#Thiet lap thing so khoi luong gao
rice ['VL'] =Fuzz.trimf(rice.universe,[0.2,0.25,0.3])
rice ['L'] =Fuzz.trimf(rice.universe,[0.25,0.3,0.5])
rice ['AV'] =Fuzz.trimf(rice.universe,[0.4,0.5,0.6])
rice ['M'] =Fuzz.trimf(rice.universe,[0.55,0.7,0.8])
rice ['VM'] =Fuzz.trimf(rice.universe,[0.75,0.9,1])
rice ['VL'].view()
```

```python
#Thiet lap thong so ve thoi gian
temp ['VL'] =Fuzz.trimf(temp.universe,[0.1,15,40])
temp ['L'] =Fuzz.trimf(temp.universe,[35,45,55])
temp ['AV'] =Fuzz.trimf(temp.universe,[52,60,70])
temp ['H'] =Fuzz.trimf(temp.universe,[65,75,85])
temp ['VH'] =Fuzz.trimf(temp.universe,[80,90,100])
temp ['VL'].view()
```

```python
#Thiet lap luat
rule1=ctrl.Rule(time['VS']&rice['VL'],temp['VL'])
rule2=ctrl.Rule(time['VS']&rice['L'],temp['VL'])
rule3=ctrl.Rule(time['VS']&rice['AV'],temp['VL'])
rule4=ctrl.Rule(time['VS']&rice['M'],temp['L'])
rule5=ctrl.Rule(time['VS']&rice['VM'],temp['L'])
rule6=ctrl.Rule(time['S']&rice['VL'],temp['VL'])
rule7=ctrl.Rule(time['S']&rice['L'],temp['L'])
rule8=ctrl.Rule(time['S']&rice['AV'],temp['L'])
rule9=ctrl.Rule(time['S']&rice['M'],temp['AV'])
rule10=ctrl.Rule(time['S']&rice['M'],temp['AV'])
rule11=ctrl.Rule(time['AV']&rice['VL'],temp['L'])
rule12=ctrl.Rule(time['AV']&rice['L'],temp['AV'])
rule13=ctrl.Rule(time['AV']&rice['AV'],temp['AV'])
rule14=ctrl.Rule(time['AV']&rice['M'],temp['H'])
rule15=ctrl.Rule(time['AV']&rice['VM'],temp['H'])
rule16=ctrl.Rule(time['F']&rice['VL'],temp['H'])
rule17=ctrl.Rule(time['F']&rice['L'],temp['H'])
rule18=ctrl.Rule(time['F']&rice['AV'],temp['H'])
rule19=ctrl.Rule(time['F']&rice['M'],temp['VH'])
rule20=ctrl.Rule(time['F']&rice['VM'],temp['VH'])
rule21=ctrl.Rule(time['VF']&rice['VL'],temp['H'])
rule22=ctrl.Rule(time['VF']&rice['L'],temp['H'])
rule23=ctrl.Rule(time['VF']&rice['AV'],temp['VH'])
rule24=ctrl.Rule(time['VF']&rice['M'],temp['VH'])
rule25=ctrl.Rule(time['VF']&rice['VM'],temp['VH'])
rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
rule10.view()
rule11.view()
rule12.view()
rule13.view()
rule14.view()
rule15.view()
rule16.view()
rule17.view()
rule18.view()
rule19.view()
rule20.view()
rule21.view()
rule22.view()
rule23.view()
rule24.view()
rule25.view()
```

```python
temping_ctrl = ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
temping = ctrl.ControlSystemSimulation(temping_ctrl)
temping.input['rice'] = 0.8
temping.input['time'] = 5

temping.compute()
```

```python
temp.view(sim=temping)
```
