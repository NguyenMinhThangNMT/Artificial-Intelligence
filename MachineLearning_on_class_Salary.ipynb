```python
import pandas as pd
url='Salary_Data.csv'
data=pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show()
```

```python
correlation =data.corr()
print(correlation)
```

```python
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black')
```

```python
import seaborn as sns
sns.heatmap(correlation,annot=True)
sns.pairplot(data)
```

```python
#su dung thu vien 
from sklearn import linear_model
model=linear_model.LinearRegression()
#Ham day hoc
model.fit(X=pd.DataFrame(data['YearsExperience']),y=data['Salary'])
print(model.intercept_)
print(model.coef_)
```

```python
model.score(X=pd.DataFrame(data['YearsExperience']),y=data['Salary'])
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['YearsExperience']))
plt.plot(data['YearsExperience'],y_pred,color='blue')
```

```python
model=linear_model.LinearRegression()
X= pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2]).T
model.fit(X,y=data['Salary'])
print('Intercept cua mon hinh',model.intercept_)
print('Coffient cua mo hinh',model.coef_)
print('Do chinh xac cua mo hinh',model.score(X,y=data['Salary']))
```

```python
import numpy as np
range=np.arange(1.5,5.5,0.1)
data.plot(kind='scatter',x='YearsExperience',y='Salary',figsize=(9,9),color='black',xlim=(0,7))
predictor=pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
